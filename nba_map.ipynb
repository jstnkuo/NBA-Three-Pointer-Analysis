```python
import pandas as pd 
import matplotlib.pyplot as plt
import hvplot.pandas
from config import api_key
import json
import requests
```

```python
place_path = "Resources/nba_arena.csv"
data = pd.read_csv(place_path)
data
```

```python
lon = []
lat = []
data["lon"] = ""
data["lat"] = ""
for i in range(30):
    base_url = "https://api.geoapify.com/v1/geocode/search?"
    params = {
        "apiKey": api_key,
        "text" : data["Arena Name"][i] + data["Arena Location"][i]}
    
    info = requests.get(base_url, params=params).json()
    #print(json.dumps(information, indent=4, sort_keys = True))
    lon.append(info["features"][0]["properties"]["lon"])
    lat.append(info["features"][0]["properties"]["lat"])

    
    
    
```

```python
data["lon"] = lon
data["lat"] = lat
```

```python
data["Opening Year"] = data["Opening Year"].astype("int")
data
```

```python
team_map = data.hvplot.points(
                    "lon",
                    "lat",
                     geo = True,
                     tiles="OSM",
                     color = "Team Name",
                      hover_cols= ["Seating Capacity"],
                    frame_width = 4000, 
                    frame_height = 3800)



team_map 
```
